```python
import matplotlib.pyplot as plt
import numpy as np
N = 3000
Rmin = 0.1
Rmax = 150
plt.figure(figsize=(18,12))
def g(R): #
   #if x <= 0: return np.exp(-x)
   #else: return np.cos(8*x**2)
    return 1/(2*np.sqrt(2*np.pi)*0.29035*R)*np.exp(-1/2*(np.log(R)-4.062)**2/0.29035**2)
p = np.linspace( Rmin, Rmax, N)
plt.plot( p, [g(p[k]) for k in range(N)] )
plt.show()
```

```python
D50=187
D84=230
lnsigma=np.log(D84/D50)
#lnD50=np.log(D50)/np.exp(3*(lnsigma)**2)
lnD50=np.log(D50) #mass
print(lnsigma)
print(lnD50)
title='g(R) [mass] (D50='+str(int(D50))+',D84='+str(int(D84))+')'
#title='test'
print(title)
```

```python
N = 3000
Rmin = 0.1
Rmax = 400
def f(R): #
    return 1/(2*np.sqrt(2*np.pi)*lnsigma*R)*np.exp(-1/2*(np.log(R)-lnD50)**2/lnsigma**2)
p = np.linspace( Rmin, Rmax, N)
plt.figure(figsize=(15,10), facecolor="white")

plt.xlabel("R(um)", fontsize=18)
plt.ylabel("g(R)", fontsize=18)
plt.title(title, fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.ylim(0,1.1)
plt.xlim(0,Rmax)
plt.plot( p, [f(p[k]) for k in range(N)] )

#plt.legend()
#plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=17)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
plt.savefig(title+".png",facecolor="white")

plt.show()
```
