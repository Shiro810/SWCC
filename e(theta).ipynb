```python
import matplotlib.pyplot as plt
import numpy as np
N = 3000
ymin = -np.pi
ymax = np.pi

plt.figure(figsize=(18,12))
k=12
Vmin=0.694
emacro=0.500
V0=np.pi/6.0*(emacro+1.0)
Vs=np.pi/6.0
def f(y): #
    return 6*(np.sin(2*y))**2/np.pi-1.0
p = np.linspace(ymin, ymax, N)
title='e(y)'
print(title)
plt.figure(figsize=(15,10), facecolor="white")

plt.xlabel("y", fontsize=20)
plt.ylabel("e=f(y)", fontsize=20)
plt.title(title, fontsize=20)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
#plt.ylim(0,)
plt.xlim(ymin,ymax)
plt.plot( p, [f(p[k]) for k in range(N)] )

#plt.legend()
#plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=17)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
plt.savefig(title+".png",facecolor="white")

plt.show()
```
