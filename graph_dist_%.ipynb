```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path='C:/Users/nobun/Dropbox/研究/00_卒論/Python_code'
#inputfolderpath=path+'/loose/dist_loose_dry_%_vtxt'
#outputfolderpath=path+'/loose/graph_loose_dry_%/'
inputfolderpath=path+'/dense/dist_dense_dry_%_vtxt/'
outputfolderpath=path+'/dense/graph_dense_dry_%/'
os.chdir(inputfolderpath)
os.makedirs(outputfolderpath,exist_ok=True)
```

```python
path_list = glob.glob("./*.txt")
print(path_list)
```

```python
#x1= list(range(1,100,2))
#x= pd.DataFrame(x1)
x = np.arange(1,100,2)
#print(x)
```

```python
for i in range(len(path_list)):
        num=str(i)
        numi=str(i*200)
        numi2=str((i+1)*200)
       
        tytle='pore volume='+numi+'-'+numi2+'  voxels'
        
        df=pd.read_csv(glob.glob("./*.txt")[i], sep='\t',header=None,engine='python')
        
        plt.figure(figsize=(15,10))
        fig, ax = plt.subplots()
        #print(df)
        sum=df.sum(axis=1)
        #print(sum)
        for j in range(1,len(df)):
            dfj=df.iloc[j]
            dfj_y=dfj+10*(j-1)
            
            plt.plot(x,dfj_y)
            ax.fill_between(x,0+10*(j-1),dfj_y,  alpha=0.5)
            #plt.grid(True)
        plt.ylim(0,160)
        plt.xlim(0,100)
        plt.legend(["a","b","c","d","e","f","g"],frameon=False,bbox_to_anchor=(0.98, 1), loc='upper right')
        #plt.legend(["c","d","e","g","i","j","k"],frameon=False,bbox_to_anchor=(0.98, 1), loc='upper right')
        #plt.legend(["c","d","e","g"],frameon=False,bbox_to_anchor=(0.98, 1), loc='upper right')
        ax.set_xlabel("Pore saturation(%)", size=15)
        ax.set_ylabel("Relative frequency(%)", size=15)
        ax.set_title(tytle)
        
        outfile=outputfolderpath+num.zfill(2)+'v_%_loose'+numi+'-'+numi2+'.png'
        fig.savefig(outfile)
```
