```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from math import gamma
gamma_u = np.frompyfunc(gamma, 1, 1)
```

```python
k=12
Vmin1=1.81
emacro=0.637
V1=4.0*np.pi/3.0*emacro
#Vs=np.pi/6.0
def f1(x): #
   #if x <= 0: return np.exp(-x)
    #return Vs*k**k/gamma_u(k)*(Vs*(x+1)-Vmin)**(k-1)/(V0-Vmin)**k*np.exp((-k*(Vs*(x+1)-Vmin)/(V0-Vmin)))
    #return 24.0/np.pi*np.sin(2.0*x)*np.cos(2.0*x)*k**k/gamma_u(k)*((np.sin(2.0*x))**2-Vmin)**(k-1)/(V0-Vmin)**k*np.exp((-k*(np.sin(2.0*x))**2-Vmin)/(V0-Vmin))
    return 32.0*np.sin(2.0*x)*np.cos(2.0*x)*k**k/gamma_u(k)*(8*(np.sin(2.0*x))**2-4.0*np.pi/3.0-Vmin1)**(k-1)/(V1-Vmin1)**k*np.exp((-k*(8*(np.sin(2.0*x))**2-4.0*np.pi/3.0-Vmin1)/(V1-Vmin1)))
```

```python
V0=np.pi/6.0*(emacro+1.0)
Vs=np.pi/6.0
Vmin0=0.694
def f0(x): #
    return 1/(5.5123196049127075e-22)*24.0/np.pi*np.sin(2.0*x)*np.cos(2.0*x)*k**k/gamma_u(k)*((np.sin(2.0*x))**2-Vmin0)**(k-1)/(V0-Vmin0)**k*np.exp((-k*(np.sin(2.0*x))**2-Vmin0)/(V0-Vmin0))
```

```python
D50=187.0
D84=230.0
lnsigma=np.log(D84/D50)
lnD50=np.log(D50)/np.exp(3*(lnsigma)**2)
print(lnsigma)
print(lnD50)
def g(R): #
    return 1/(2*np.sqrt(2*np.pi)*lnsigma*R)*np.exp(-1/2*(np.log(R)-lnD50)**2/lnsigma**2)
```

```python
Vv1= lambda R, x: (8*(np.sin(2.0*x))**2-4.0/3.0*np.pi)*R**3*f1(x)*g(R)
Vtotal1=integrate.dblquad(Vv1, np.pi/6.0, np.pi/4.0, lambda x: 0, lambda x: 10000)
print(Vtotal1)
T=72.0 #(kN/um)

def Vsat_total1(s) :
    return integrate.dblquad(Vv1, np.pi/6.0, np.pi/4.0, lambda x: 0, lambda x: 2*T*np.cos(x)/((1-np.cos(x))*s))
```

```python
Vv0= lambda R, x: (8*(np.sin(2.0*x))**2-4.0/3.0*np.pi)*R**3*f0(x)*g(R)
Vtotal0=integrate.dblquad(Vv0, np.pi/6.0, np.pi/4.0, lambda x: 0, lambda x: 10000)
print(Vtotal0)
T=72.0 #(kN/um)

def Vsat_total0(s) :
    return integrate.dblquad(Vv0, np.pi/6.0, np.pi/4.0, lambda x: 0, lambda x: 2*T*np.cos(x)/((1-np.cos(x))*s))
```
