```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path='C:/Users/nobun/Dropbox/研究/00_卒論/Python_code'
#inputfolderpath=path+'/loose/dist_loose_dry_%_vtxt'
#outputfolderpath=path+'/loose/graph_loose_dry_%/'
inputfolderpath=path+'/dense/dist_dense_dry_vtxt/'
outputfolderpath=path+'/dense/graph_dense_dry_voxels/'
os.chdir(inputfolderpath)
os.makedirs(outputfolderpath,exist_ok=True)
```

```python
path_list = glob.glob("./*.txt")
print(path_list)
```

```python
#x1= list(range(1,100,2))
#x= pd.DataFrame(x1)
x = np.arange(1,100,2)
#print(x)
```

```python
for i in range(len(path_list)):
        num=str(i)
        numi=str(i*200)
        numi2=str((i+1)*200)
        tytle='pore volume='+numi+'-'+numi2+' voxels'
        
        df=pd.read_csv(glob.glob("./*.txt")[i], sep='\t',header=None,engine='python')
        #dft=df.T
        plt.figure(figsize=(15,10))
        fig, ax = plt.subplots()
        #print(df)
        
        for j in range(1,len(df)):
            dfj=df.iloc[j]
            dfj_y=dfj+1000*(j-1)
            
            plt.plot(x,dfj_y)
            ax.fill_between(x,0+1000*(j-1),dfj_y,  alpha=0.5)
        plt.ylim(0,16000)
        plt.xlim(0,100)
        plt.legend(["a","b","c","d","e","f","g"],frameon=False,bbox_to_anchor=(0.98, 1), loc='upper right')

        #plt.legend(["c","d","e","g","i","j","k"])
        #plt.legend(["c","d","e","g"])
        ax.set_xlabel("Sr(%)")
        ax.set_ylabel("Number of pores")
        ax.set_title(tytle)
        outfile=outputfolderpath+num.zfill(2)+'v_number_dense'+numi+'-'+numi2+'.png'
        fig.savefig(outfile)
```
