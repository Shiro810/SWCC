```python
import matplotlib.pyplot as plt
import numpy as np
from scipy import integrate
from math import gamma
gamma_u = np.frompyfunc(gamma, 1, 1)
```

```python
T=72.0 #(kN/um)
```

```python
k=12
Vmin=0.694
emacro=0.673
V0=np.pi/6.0*(emacro+1.0)
Vs=np.pi/6.0
def f(x): #
    return 1/(5.5123196049127075e-22)*24.0/np.pi*np.sin(2.0*x)*np.cos(2.0*x)*k**k/gamma_u(k)*((np.sin(2.0*x))**2-Vmin)**(k-1)/(V0-Vmin)**k*np.exp((-k*(np.sin(2.0*x))**2-Vmin)/(V0-Vmin))
```

```python
def g(R): #
    return 1/(2*np.sqrt(2*np.pi)*0.29035*R)*np.exp(-1/2*(np.log(R)-4.062)**2/0.29035**2)
```

```python
R=100
def Vcap(s):
    return 32*np.pi*R*T**2/s**2*(1-np.arcsin(s*R/(4*T+s*R))*(np.sqrt(16*T**2+8*T*s*R))/(s*R))
```

```python
N = 100
smin = 0.1 #(kPa)
smax = 100 #(kPa)

p = np.linspace(smin, smax, N)
plt.figure(figsize=(15,10), facecolor="white")
#ax = fig.add_subplot(111)
plt.xlabel("Suction(kPa)", fontsize=18)
plt.ylabel("Vcap", fontsize=18)
#ax.set_xlabel("Suction(kPa)", fontsize=18)
#ax.set_ylabel("Sr(%)", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)

plt.plot( p, [Vcap(p[k]) for k in range(N)],label='Vcap')
plt.legend()
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=17)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
plt.savefig('Vcap',facecolor="white")
```

```python
s=1000
print(Vcap(s))
```

```python
s=10
N_Vcap= lambda R, x: f(x)*g(R)
def N_Vcap_total(s) :
    return integrate.dblquad(N_Vcap, np.pi/6.0, np.pi/4.0, lambda x: 2*T*np.cos(x)/((1-np.cos(x))*s), lambda x: 1000)
```

```python
N = 50
smin = 0.1 #(kPa)
smax = 100 #(kPa)

p = np.linspace(smin, smax, N)
plt.figure(figsize=(15,10), facecolor="white")
#ax = fig.add_subplot(111)
plt.xlabel("Suction(kPa)", fontsize=18)
plt.ylabel("N_Vcap", fontsize=18)
#ax.set_xlabel("Suction(kPa)", fontsize=18)
#ax.set_ylabel("Sr(%)", fontsize=18)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.ylim(0,1.6)
plt.xlim(0,100)
plt.plot( p, [N_Vcap_total(p[k]) for k in range(N)], label='N_Vcap') #
plt.legend()
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=17)
plt.subplots_adjust(left=0.1, right=0.95, bottom=0.1, top=0.95)
plt.savefig('N_Vcap',facecolor="white")
```
