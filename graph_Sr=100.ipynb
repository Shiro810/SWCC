```python
import os
import glob
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows', None)
```

```python
path='C:/Users/nobun/Dropbox/研究/00_卒論/Python_code'
#inputfolderpath=path+'/loose/dist_loose_dry_%_vtxt'
#outputfolderpath=path+'/loose/graph_loose_dry_%/'
inputfile=path+'/dense/Sr=100_dense_dry_%/Sr=100_dense_dry.txt'
outputfolderpath=path+'/dense/graph_Sr=100_dense_dry/'

os.makedirs(outputfolderpath,exist_ok=True)
```

```python
df=pd.read_csv(inputfile, sep='\t',header=None,engine='python')
#print(df)
vji0=df.iloc[1:,0]
print(vji0)
print(len(df.columns))
```

```python
for i in range(len(df.columns)):
    num=str(i)
    numi=str(i*200)
    numi2=str((i+1)*200)
       
    plt.figure(figsize=(15,10))
    fig, ax = plt.subplots()
    
    left = np.array([1, 2, 3, 4, 5, 6, 7])
    height = df.iloc[1:,i]
    label = ["a","b","c","d","e","f","g"]
    plt.ylim(0,100)
    
    
    plt.bar(left, height,tick_label=label,color='lightgrey',edgecolor="dimgrey",linewidth=1)
    
    tytle='pore volume='+numi+'-'+numi2+' voxels'
    ax.set_ylabel("Proportion of saturated pores (%)")
    ax.set_title(tytle)
    outfile=outputfolderpath+num.zfill(2)+'Sr=100_dense_dry'+numi+'-'+numi2+'.png'
    fig.savefig(outfile)
```
