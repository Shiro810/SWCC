```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from scipy.stats import gamma
def calc_gamma_param(mu, k):
    return k, 1.0/k

mu1, k1 = 1, 12.0
shape1, scale1 = calc_gamma_param(mu1, k1)
mu2, k2 = 1, 13.0
shape2, scale2 = calc_gamma_param(mu2, k2)
mu3, k3 = 1, 14.0
shape3, scale3 = calc_gamma_param(mu3, k3)
mu4, k4 = 1, 15.0
shape4, scale4 = calc_gamma_param(mu4, k4)
plt.figure(figsize=(15,10), facecolor="white")
def plot_gamma(xx, shape, scale):
    plt.plot(xx, sp.stats.gamma(a=shape, scale=scale).pdf(xx), label=f'shape={shape}, scale={scale}')
title='k-Gamma distribution'
xx = np.linspace(0.3, 2)
plot_gamma(xx, shape1, scale1)
plot_gamma(xx, shape2, scale2)
plot_gamma(xx, shape3, scale3)
plot_gamma(xx, shape4, scale4)
plt.title(title, fontsize=20)
plt.legend()
plt.xlabel("x", fontsize=18)
plt.ylabel("p(x)", fontsize=18)
plt.xlim(0.3,2)
plt.ylim(0,1.7)
plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.legend(bbox_to_anchor=(1, 1), loc='upper right', borderaxespad=1, fontsize=15)
plt.savefig(title+".png",facecolor="white")
```
